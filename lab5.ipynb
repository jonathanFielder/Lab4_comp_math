```python
def triangle(a, b, c):
	val = " not "
	if a + b > c and a + c > b and b + c > a:
		val = " "
	print("The numbers", str(a) + ",", str(b) + ",", str(c) + ",", "are" + val + "the sides of a valid triangle.")
```

# Question 7:

```python
def addintegers(N):
	n = 1
	sum = 0
	while sum + n <= N:
		sum +=N
		n += 1
	print("You need to add", n, "integers so that the sum is greater than", str(N) + ".")
	
```
